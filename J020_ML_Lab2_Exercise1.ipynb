```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
headers= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
 "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
 "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
 "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
d=pd.read_csv("C:\\Users\\91773\\Desktop\\auto.csv",names=headers)
```

```python
d
```

```python
d.isnull().sum()
```

```python
d.replace('?',np.nan,inplace=True)
```

```python
d.isnull().sum()
```

```python
d.drop(d[d['price'].isnull()].index,axis=0,inplace=True)
d
```

```python
r1 = d['normalized-losses'].astype('float').mean(axis =0)
d['normalized-losses'].replace(np.nan,r1,inplace = True)

r2 = d['bore'].astype('float').mean(axis =0)
d['bore'].replace(np.nan,r2,inplace = True)

r3 = d['stroke'].astype('float').mean(axis =0)
d['stroke'].replace(np.nan,r3,inplace = True)

r4 = d['horsepower'].astype('float').mean(axis =0)
d['horsepower'].replace(np.nan,r4,inplace = True)

r5 = d['peak-rpm'].astype('float').mean(axis =0)
d['peak-rpm'].replace(np.nan,r5,inplace = True)
```

```python
d
```

```python
d['num-of-doors'].replace(np.nan,d['num-of-doors'].value_counts().idxmax(),inplace = True )
d
```

```python
d.reset_index()
```

```python
for i in d._get_numeric_data().columns:
    d[i] = d[i].astype('float')
```

```python
d['city-L/100km'] = 235/d['city-mpg']
d['highway-L/100km'] = 235/d['highway-mpg']
d
```

```python
d['length'] = d['length']/d['length'].max()
d['width'] = d['width']/d['width'].max()
d['height'] = d['height']/d['height'].max()
```

```python
d['horsepower'] = d['horsepower'].astype('int')
plt.hist(d.horsepower, bins='auto', rwidth=2, histtype='bar')
```

```python
bins=pd.cut(d['horsepower'], bins = 3, labels = ['low','medium','high'])
d['horsepower-binned'] =bins
d
```

```python
plt.hist(d['horsepower-binned'])
```

```python
one_hot1 = pd.get_dummies(d['fuel-type'])
one_hot2 = pd.get_dummies(d['aspiration'])
d = d.join(one_hot1)
d = d.join(one_hot2)
d.drop("fuel-type", axis = 1, inplace=True)
d.drop("aspiration", axis = 1, inplace=True)

d
```
